```python
# prompt: crie um programa onde faça o controle de manutenções preventivas

import pandas as pd
from datetime import datetime, timedelta

# Função para adicionar uma nova manutenção
def adicionar_manutencao():
  equipamento = input("Nome do equipamento: ")
  descricao = input("Descrição da manutenção: ")
  data_programada = input("Data programada (DD-MM-AAAA): ")
  responsavel = input("Responsável pela manutenção: ")

  nova_manutencao = {
      'Equipamento': equipamento,
      'Descrição': descricao,
      'Data Programada': data_programada,
      'Responsável': responsavel,
      'Status': 'Pendente'
  }

  return nova_manutencao



 Função para listar as manutenções
def listar_manutencoes(df_manutencoes):
  print("\nManutenções:")
df=pd.DataFrame(df_manutencoes)
df

# Função para marcar uma manutenção como concluída
def concluir_manutencao(df_manutencoes):
  listar_manutencoes(df_manutencoes)
  indice_manutencao = int(input("Digite o índice da manutenção a ser concluída: "))

  if 0 <= indice_manutencao < len(df_manutencoes):
      df_manutencoes.loc[indice_manutencao, 'Status'] = 'Concluída'
      print("Manutenção marcada como concluída!")
  else:
      print("Índice inválido.")

  return df_manutencoes


# Inicialização do DataFrame de manutenções (opcional: carregar de um arquivo CSV)
df_manutencoes = pd.DataFrame(columns=['Equipamento', 'Descrição', 'Data Programada', 'Responsável', 'Status'])


while True:
    print("\nOpções:")
    print("1. Adicionar nova manutenção")
    print("2. Listar manutenções")
    print("3. Marcar manutenção como concluída")
    print("4. Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == '1':
        nova_manutencao = adicionar_manutencao()
        df_manutencoes = pd.concat([df_manutencoes, pd.DataFrame([nova_manutencao])], ignore_index=True)
    elif opcao == '2':
        listar_manutencoes(df_manutencoes)
    elif opcao == '3':
        df_manutencoes = concluir_manutencao(df_manutencoes)
    elif opcao == '4':
        break
    else:
        print("Opção inválida.")
```
